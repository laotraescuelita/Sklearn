<h1> Mostrar las diferentes librerias para aplicar algoritmos de aprendizaje automático </h1>

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Craer datos de simulación.
from sklearn.datasets import make_regression
features, target, coefficients = make_regression(n_samples = 100,
    n_features = 3,
    n_informative = 3,
    n_targets = 1,
    noise = 0.0,
    coef = True,
    random_state = 1)
print('Matriz \n', features[:3])
print('Vector \n', target[:3])
```

```python
#Dividir los datos para entrenamiento y prueba.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42)

#Escojer un módelo
from sklearn.linear_model import LinearRegression

#Entrenar el módelo.
modelo = LinearRegression().fit(X_train, y_train)

print("Coeficientes: {}".format(modelo.coef_))
print("Intecepto: {}".format(modelo.intercept_))

print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
```

<h1>Evaluemos otros modelos </h1>

```python
from sklearn.linear_model import Ridge
modelo = Ridge().fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
```

```python
modelo = Ridge(alpha=10).fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
```

```python
modelo = Ridge(alpha=0.1).fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
```

```python
from sklearn.linear_model import Lasso
modelo = Lasso().fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
print("Número de varibles utilizadas: {}".format(np.sum(modelo.coef_ != 0)))
```

```python
# we increase the default setting of "max_iter",
# otherwise the model would warn us that we should increase max_iter.
modelo = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
print("Número de varibles utilizadas: {}".format(np.sum(modelo.coef_ != 0)))
```

```python
modelo = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
print("Número de varibles utilizadas: {}".format(np.sum(modelo.coef_ != 0)))
```

```python
from sklearn.tree import DecisionTreeRegressor
modelo = DecisionTreeRegressor().fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
```

```python
from sklearn.svm import SVR
modelo = SVR().fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
```

```python
from sklearn import linear_model
modelo = linear_model.BayesianRidge().fit(X_train, y_train)
print("Porcentaje de acierto entrenamiento: {:.2f}".format(modelo.score(X_train, y_train)))
print("Porcentaje de acierto pruebas: {:.2f}".format(modelo.score(X_test, y_test)))
```
